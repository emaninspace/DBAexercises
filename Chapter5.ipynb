```r
# Exercise 5.1
# Let theta denote the true presence of a disease so theta = 1 is have that disease and theta = 0 is not have the disease
# p( theta = 1 ) =  0.001 and p( theta = 0 ) = 0.999
# Let T = 1 denote a positive test result (has the disease) and T = 0 denote a negative test result (does not have the disease)
# P( T = 1 | theta = 1 ) = 0.99 and P( T = 1 | theta = 0 ) = 0.05

pTheta <- c(0.001, 0.999)

pTgivenTheta <- matrix(c(0.99, 0.01, .05, 0.95), nrow=2, ncol=2,
                           dimnames = list( RowDimName = c('T=+','T=-'),
                                            ColDimName = c('+','-')))

for(i in 1:2) {
    # Compute the conditional matrix using Bayes' theorem
    expandedMatrix <- c( pTgivenTheta[1,1]*pTheta[1]/( pTgivenTheta[1,1]*pTheta[1] + pTgivenTheta[1,2]*pTheta[2]),
                         pTgivenTheta[1,2]*pTheta[2]/( pTgivenTheta[1,1]*pTheta[1] + pTgivenTheta[1,2]*pTheta[2]),
                         pTgivenTheta[2,1]*pTheta[1]/( pTgivenTheta[2,1]*pTheta[1] + pTgivenTheta[2,2]*pTheta[2]),
                         pTgivenTheta[2,2]*pTheta[2]/( pTgivenTheta[2,1]*pTheta[1] + pTgivenTheta[2,2]*pTheta[2]))
    pThetaGivenT <- matrix( expandedMatrix,
                           nrow = 2, ncol = 2,
                           dimnames = list( RowDimName = c('+','-'),
                                            ColDimName = c('T=+','T=-')
                                          ) 
                          )
    pTheta <- c( pThetaGivenT[1,1], 1-pThetaGivenT[1,1])
    if(i == 1) {
        show(pThetaGivenT[1,1])
    }
    if(i == 2) {
        show(pThetaGivenT[1,2])
    }
}
```
